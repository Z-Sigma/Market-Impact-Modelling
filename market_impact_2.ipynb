```python
import pandas as pd
import numpy as np

# === CONFIGURATION ===
TOP_LEVELS = 10
NUM_SIMULATED_TRADES = 5000  # total number of samples
TRADE_SIZE_PROPORTIONS = (0.4, 0.35, 0.25)  # small, medium, large proportions

# === STEP 1: CLEAN AND CONVERT TIMESTAMP ===
def preprocess_combined_df(combined_df):
    combined_df['ts_event'] = pd.to_datetime(combined_df['ts_event'], format='mixed')
    combined_df['minute'] = combined_df['ts_event'].dt.floor('min')

    columns_to_drop = [
        'ts_event.1', 'rtype', 'publisher_id', 'instrument_id', 'action', 'side',
        'depth', 'price', 'size', 'flags', 'ts_in_delta', 'sequence'
    ]
    combined_df.drop(columns=[c for c in columns_to_drop if c in combined_df.columns], inplace=True)
    return combined_df

# === STEP 2: GET MINUTE-LEVEL SNAPSHOTS (MEAN) ===
def get_minute_snapshots_mean(df):
    df = df.copy()
    df['minute'] = df['ts_event'].dt.floor('min')
    feature_cols = [f'{side}_{field}_0{i}' for i in range(TOP_LEVELS)
                    for side in ['bid', 'ask'] for field in ['px', 'sz', 'ct']]
    feature_cols = [col for col in feature_cols if col in df.columns]
    grouped = df.groupby(['minute', 'symbol'])[feature_cols].mean().reset_index()
    return grouped

# === STEP 3: FEATURE ENGINEERING ===
def compute_features(df):
    df = df.copy()
    df['mid_price'] = (df['bid_px_00'] + df['ask_px_00']) / 2
    df['spread'] = df['ask_px_00'] - df['bid_px_00']
    df['imbalance'] = df['bid_sz_00'] / (df['bid_sz_00'] + df['ask_sz_00'] + 1e-6)

    df['cum_bid_size'] = sum([df[f'bid_sz_0{i}'] for i in range(TOP_LEVELS)])
    df['cum_ask_size'] = sum([df[f'ask_sz_0{i}'] for i in range(TOP_LEVELS)])
    df['cum_bid_count'] = sum([df[f'bid_ct_0{i}'] for i in range(TOP_LEVELS)])
    df['cum_ask_count'] = sum([df[f'ask_ct_0{i}'] for i in range(TOP_LEVELS)])

    df['bid_width'] = df['bid_px_00'] - df['bid_px_09']
    df['ask_width'] = df['ask_px_09'] - df['ask_px_00']
    df['bid_price_slope'] = df['bid_width'] / (df['cum_bid_size'] + 1e-6)
    df['ask_price_slope'] = df['ask_width'] / (df['cum_ask_size'] + 1e-6)

    df['size_asymmetry'] = df['cum_bid_size'] - df['cum_ask_size']
    df['count_asymmetry'] = df['cum_bid_count'] - df['cum_ask_count']

    df['minute_of_day'] = df['minute'].dt.hour * 60 + df['minute'].dt.minute
    df['sin_time'] = np.sin(2 * np.pi * df['minute_of_day'] / 390)
    df['cos_time'] = np.cos(2 * np.pi * df['minute_of_day'] / 390)

    return df

# === STEP 4: SLIPPAGE SIMULATION FUNCTION ===
def simulate_slippage(row, x):
    total_cost = 0
    shares_left = x
    for i in range(TOP_LEVELS):
        price = row.get(f'ask_px_0{i}', np.nan)
        size = row.get(f'ask_sz_0{i}', 0)
        if np.isnan(price) or shares_left <= 0:
            break
        trade_size = min(size, shares_left)
        total_cost += trade_size * price
        shares_left -= trade_size

    if shares_left > 0:
        return np.nan
    avg_exec_price = total_cost / x
    return avg_exec_price - row['mid_price']

# === STEP 5: GENERATE GT DATASET WITH BALANCED TRADE SIZE MIXTURE ===
def generate_gt_dataset_mixed(features_df, num_samples=NUM_SIMULATED_TRADES, proportions=TRADE_SIZE_PROPORTIONS):
    n_small = int(num_samples * proportions[0])
    n_medium = int(num_samples * proportions[1])
    n_large = num_samples - n_small - n_medium

    sampled = features_df.sample(n=num_samples, replace=True).copy()

    x_small = np.random.randint(50, 300, size=n_small)
    x_medium = np.random.randint(300, 800, size=n_medium)
    x_large = np.random.randint(800, 1500, size=n_large)

    x_all = np.concatenate([x_small, x_medium, x_large])
    np.random.shuffle(x_all)

    sampled['x'] = x_all
    sampled['slippage'] = sampled.apply(lambda row: simulate_slippage(row, row['x']), axis=1)

    return sampled.dropna(subset=['slippage'])

# === STEP 6: FINALIZE CLEANED DATASET ===
def finalize_dataset(df):
    return df[[
        'minute', 'symbol', 'x', 'mid_price', 'spread', 'imbalance',
        'cum_bid_size', 'cum_ask_size', 'cum_bid_count', 'cum_ask_count',
        'bid_width', 'ask_width', 'bid_price_slope', 'ask_price_slope',
        'size_asymmetry', 'count_asymmetry', 'sin_time', 'cos_time',
        'slippage'
    ]]
```

```python

# --- CONFIGURATION ---
DATASET_PATH = "/content/shuffled_df.csv"
NUM_BINS = 25 # The number of buckets to group trade sizes into for the trend line

# --- LOAD THE DATASET ---
try:
    df = pd.read_csv(DATASET_PATH)
    print(f"✅ Successfully loaded dataset with {len(df)} rows.")
except FileNotFoundError:
    print(f"❌ Error: The file was not found at {DATASET_PATH}")
    print("Please make sure the file path is correct.")
    df = None

if df is not None:
    # --- CALCULATE THE TREND LINE ---
    # 1. Create bins (buckets) for the trade size 'x'
    df['x_bin'] = pd.cut(df['x'], bins=NUM_BINS)

    # 2. Calculate the average slippage and average trade size for each bin
    binned_trend = df.groupby('x_bin', as_index=False).agg(
        avg_slippage=('slippage', 'mean'),
        avg_x=('x', 'mean')
    )
    # The 'as_index=False' is deprecated in some versions, so we handle both cases
    if 'x_bin' in binned_trend.columns:
         binned_trend = df.groupby('x_bin').agg(
            avg_slippage=('slippage', 'mean'),
            avg_x=('x', 'mean')
        ).reset_index()


    print("\n--- Trend Data ---")
    print(binned_trend.head())

    # --- PLOT THE RESULTS ---
    plt.style.use('seaborn-v0_8-whitegrid')
    plt.figure(figsize=(12, 7))

    # Layer 1: Scatter plot of all the raw data points
    plt.scatter(
        df['x'],
        df['slippage'],
        alpha=0.1, # Use transparency to see density
        label='Individual Simulated Trades',
        color='lightblue'
    )

    # Layer 2: Line plot of the binned averages (the trend)
    plt.plot(
        binned_trend['avg_x'],
        binned_trend['avg_slippage'],
        color='red',
        linewidth=2.5,
        marker='o',
        markersize=5,
        label='Average Slippage Trend'
    )

    # --- STYLING ---
    plt.title('Slippage vs. Trade Size: Underlying Trend', fontsize=16)
    plt.xlabel('Trade Size (x)', fontsize=12)
    plt.ylabel('Simulated Slippage', fontsize=12)
    plt.legend(fontsize=10)
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)

    # Set y-axis to start from 0 for better context
    plt.ylim(bottom=0)
    plt.xlim(left=0)

    plt.show()

```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler, FunctionTransformer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import r2_score
import lightgbm as lgb
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore', category=UserWarning)

# --- CONFIGURATION ---
DATASET_PATH = "/content/shuffled_df.csv"
TEST_SET_SIZE = 0.2 # Use 20% of the data for the final evaluation

# --- LOAD THE DATASET ---
try:
    df = pd.read_csv(DATASET_PATH)
    print(f"✅ Successfully loaded dataset with {len(df)} rows.")
except FileNotFoundError:
    print(f"❌ Error: The file was not found at {DATASET_PATH}")
    df = None

if df is not None:
    # Filter out any non-positive slippage values before log transform
    df_filtered = df[df['slippage'] > 0].copy()

    # --- TRAIN-TEST SPLIT ---
    # Split the data BEFORE any fitting to ensure the test set is completely unseen.
    train_df, test_df = train_test_split(df_filtered, test_size=TEST_SET_SIZE, random_state=42)
    print(f"\nSplitting data into {len(train_df)} training samples and {len(test_df)} testing samples.")

    # --- STAGE 1: FIT FOR THE SHAPE PARAMETER (gamma) ---
    print("\n--- Running Stage 1: Fitting for gamma (γ) on TRAINING data ---")

    X_stage1_train = np.log(train_df[['x']])
    y_stage1_train = np.log(train_df['slippage'])

    model_stage1 = LinearRegression()
    model_stage1.fit(X_stage1_train, y_stage1_train)
    gamma = model_stage1.coef_[0]

    print(f"✅ Fitted Shape Parameter (gamma): {gamma:.4f}")

    # --- STAGE 2: ADVANCED MODELING FOR alpha_t ---
    print("\n--- Running Stage 2: Modeling alpha (α_t) with Hyperparameter Tuning ---")

    # Calculate log_alpha for both train and test sets using the SAME gamma
    train_df['log_alpha'] = np.log(train_df['slippage']) - gamma * np.log(train_df['x'])
    test_df['log_alpha'] = np.log(test_df['slippage']) - gamma * np.log(test_df['x'])

    # Define feature sets for selective transformation
    feature_columns = [
        'spread', 'imbalance', 'cum_bid_size', 'cum_ask_size',
        'cum_bid_count', 'cum_ask_count', 'bid_width', 'ask_width',
        'bid_price_slope', 'ask_price_slope', 'size_asymmetry',
        'count_asymmetry', 'sin_time', 'cos_time'
    ]

    log_transform_cols = [
        'spread', 'cum_bid_size', 'cum_ask_size', 'cum_bid_count',
        'cum_ask_count', 'bid_width', 'ask_width'
    ]
    standard_scale_cols = [col for col in feature_columns if col not in log_transform_cols]

    preprocessor = ColumnTransformer(
        transformers=[
            ('log_features', Pipeline(steps=[
                ('log', FunctionTransformer(np.log1p)),
                ('scaler', StandardScaler())
            ]), log_transform_cols),
            ('standard_features', StandardScaler(), standard_scale_cols)
        ],
        remainder='passthrough'
    )

    # --- HYPERPARAMETER TUNING on TRAINING data ---
    print("\n--- Performing Hyperparameter Tuning for LightGBM ---")

    param_grid = {
        'regressor__n_estimators': [100, 200, 300],
        'regressor__learning_rate': [0.05, 0.1, 0.2],
        'regressor__num_leaves': [31, 50, 70],
    }

    pipeline_for_tuning = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('regressor', lgb.LGBMRegressor(random_state=42, verbosity=-1))
    ])

    X_train = train_df[feature_columns]
    y_train = train_df['log_alpha']

    grid_search = GridSearchCV(pipeline_for_tuning, param_grid, cv=3, n_jobs=-1, scoring='r2', verbose=1)
    grid_search.fit(X_train, y_train)

    print(f"\n✅ Best R² score from tuning (on train data): {grid_search.best_score_:.4f}")
    print(f"✅ Best parameters found: {grid_search.best_params_}")

    best_pipeline = grid_search.best_estimator_

    # --- FINAL EVALUATION on the TEST SET ---
    print("\n--- Evaluating Final Tuned Model on UNSEEN TEST DATA ---")

    X_test = test_df[feature_columns]
    y_test_log_alpha = test_df['log_alpha']
    y_test_slippage = test_df['slippage']

    # Predict log_alpha on the test set
    log_alpha_pred_test = best_pipeline.predict(X_test)

    # Calculate R² score for the log_alpha prediction
    score_log_alpha = r2_score(y_test_log_alpha, log_alpha_pred_test)
    print(f"✅ R² Score for log(alpha) prediction on TEST SET: {score_log_alpha:.4f}")

    # Predict total slippage on the test set
    total_slippage_pred_test = np.exp(log_alpha_pred_test) * (test_df['x']**gamma)

    # Calculate R² score for the final slippage prediction
    score_slippage = r2_score(y_test_slippage, total_slippage_pred_test)
    print(f"✅ R² Score for total slippage prediction on TEST SET: {score_slippage:.4f}")

    # --- PLOT THE RESULTS on the TEST SET ---
    plt.style.use('seaborn-v0_8-whitegrid')
    plt.figure(figsize=(12, 7))
    plt.scatter(y_test_slippage, total_slippage_pred_test, alpha=0.2, label='Model Predictions on Test Set')
    plt.plot([0, y_test_slippage.max()], [0, y_test_slippage.max()], color='red', linestyle='--', label='Perfect Fit')
    plt.title('Predicted vs. Actual Slippage (on Test Set)', fontsize=16)
    plt.xlabel('Actual Slippage', fontsize=12)
    plt.ylabel('Predicted Slippage', fontsize=12)
    plt.legend()
    plt.grid(True)
    plt.xlim(0, np.quantile(y_test_slippage, 0.99))
    plt.ylim(0, np.quantile(total_slippage_pred_test, 0.99))
    plt.show()

```

This pipeline models the temporary price impact, gt = αt * x^γ
 , by breaking the problem into two distinct stages:

Stage 1: Find the Universal "Shape" (γ)
The first step assumes that the fundamental shape of the price impact curve is constant across all market conditions. It uses a simple linear regression on the log of slippage and the log of trade size (x) to find a single, robust value for the exponent gamma. This parameter defines the concavity of the impact curve.

Stage 2: Predict the Dynamic "Level" (αt)
The second step models how the overall level of the impact curve moves up or down based on current market liquidity. It first calculates the alpha for each data point using the gamma from Stage 1. Then, it trains a powerful, non-linear LightGBM model to predict this alpha using all the other market features (like spread, imbalance, etc.). This stage uses hyperparameter tuning (GridSearchCV) and selective feature transformations to ensure the model is as accurate and robust as possible.

In short, the pipeline first finds the shape of the impact curve and then predicts its level for any given moment, resulting in a complete and highly predictive model of total slippage.
